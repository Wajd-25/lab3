```python
import numpy as np
import matplotlib.pyplot as plt
from skimage import io
from google.colab import files

uploaded = files.upload()
image_name = list(uploaded.keys())[0]

image = io.imread(image_name, as_gray=True)
image = (image * 255).astype(np.uint8)
```

```python
# Exercise 2.1

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title("Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.hist(image.flatten(), bins=256, range=(0,255))
plt.title("Histogram")
plt.xlabel("Grey level")
plt.ylabel("Frequency")

plt.show()
```

```python
# Exercise 3.1

hist, bins = np.histogram(image.flatten(), bins=256, range=(0,255))
hist_norm = hist / hist.sum()

print(hist_norm.sum())
```

```python
# Exercise 4.1

def hist_equalise(img):
    hist, _ = np.histogram(img.flatten(), bins=256, range=(0,255))
    cdf = hist.cumsum()
    cdf_norm = (cdf - cdf.min()) * 255 / (cdf.max() - cdf.min())
    cdf_norm = cdf_norm.astype(np.uint8)
    return cdf_norm[img]

equalised = hist_equalise(image)

plt.figure(figsize=(10,4))

plt.subplot(1,2,1)
plt.imshow(equalised, cmap='gray')
plt.title("Equalised Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.hist(equalised.flatten(), bins=256, range=(0,255))
plt.title("Equalised Histogram")

plt.show()
```

```python
# Exercise 7.1

threshold = 128

binary = np.zeros_like(image)
binary[image > threshold] = 255

plt.imshow(binary, cmap='gray')
plt.title("Thresholded Image")
plt.axis('off')
plt.show()
```

```python
# Exercise 8.1

image_norm = image / 255.0
c = 1 / np.log(1 + np.max(image_norm))
log_img = c * np.log(1 + image_norm)
log_img = np.uint8(log_img * 255)

plt.imshow(log_img, cmap='gray')
plt.title("Log Transformed Image")
plt.axis('off')
plt.show()
```

```python
# Exercise 8.2

gamma = 0.4

image_norm = image / 255.0
gamma_img = np.power(image_norm, gamma)
gamma_img = np.uint8(gamma_img * 255)

plt.imshow(gamma_img, cmap='gray')
plt.title("Gamma Correction ( = 0.4)")
plt.axis('off')
plt.show()
```

```python
# Exercise 10.1

from scipy.ndimage import laplace

lap = laplace(image)
sharpened = image - lap
sharpened = np.clip(sharpened, 0, 255).astype(np.uint8)

plt.imshow(sharpened, cmap='gray')
plt.title("Laplacian Sharpening")
plt.axis('off')
plt.show()
```

```python
# Exercise 10.2

from scipy.ndimage import sobel

sx = sobel(image, axis=0)
sy = sobel(image, axis=1)

edges = np.hypot(sx, sy)
edges = np.clip(edges, 0, 255).astype(np.uint8)

plt.imshow(edges, cmap='gray')
plt.title("Sobel Edge Magnitude")
plt.axis('off')
plt.show()
```

```python
# Exercise 12  Mini Project

# Step 1: Histogram Equalisation
def hist_equalise(img):
    hist, _ = np.histogram(img.flatten(), bins=256, range=(0,255))
    cdf = hist.cumsum()
    cdf_norm = (cdf - cdf.min()) * 255 / (cdf.max() - cdf.min())
    cdf_norm = cdf_norm.astype(np.uint8)
    return cdf_norm[img]

equalised = hist_equalise(image)

# Step 2: Smoothing (Mean Filter)
from scipy.ndimage import uniform_filter
smoothed = uniform_filter(equalised, size=3)

# Step 3: Sharpening (Laplacian)
from scipy.ndimage import laplace
lap = laplace(smoothed)
final = smoothed - lap

# Keep values within valid range
final = np.clip(final, 0, 255).astype(np.uint8)

# Step 4: Compare Original and Final Images
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(final, cmap='gray')
plt.title("Final Enhanced Image")
plt.axis('off')

plt.show()
```
